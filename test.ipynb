## Dataset error detection helper  
This section adds a reusable `detect_errors` utility that inspects a DataFrame for common issues:  
- Missing values summary (per column and percent)  
- Data type inconsistencies and coercion suggestions  
- Date parsing errors (unparsed/NaT counts)  
- Numeric outliers (IQR and z-score methods)  
- Duplicate rows  
- Identifier (columns ending with 'Id'/'ID') format checks  
- Column cardinality and suspicious constant or near-constant columns  
  
Usage: run the next code cell after `flood_project_df` is loaded. Call `report, details = detect_errors(flood_project_df)` and inspect `report` (summary) and `details` (dictionaries with lists/indices).

```python
def detect_errors(df, date_formats=None, id_regex=r"^[A-Za-z0-9\-_.]+$", zscore_thresh=3.5, iqr_multiplier=1.5):
    """Inspect DataFrame for common data quality issues and return a summary report and details.
    Returns:
      summary_df: pandas DataFrame with one-row-per-check summary (counts/percents)
      details: dict with detailed lists/indexes for each check

    Parameters:
      date_formats: optional dict {col: format_string} to try for specific date columns
      id_regex: regex for valid ID-like values (defaults to alnum, dash, underscore, dot)
    """
    import pandas as pd
    import numpy as np
    from scipy import stats

    n = df.shape[0]
    details = {}

    # Missing values
    miss_count = df.isnull().sum()
    miss_pct = (miss_count / n) * 100
    missing_summary = pd.DataFrame({'missing_count': miss_count, 'missing_pct': miss_pct})
    details['missing'] = {col: int(miss_count[col]) for col in df.columns if miss_count[col] > 0}

    # Duplicates
    dup_mask = df.duplicated(keep=False)
    dup_count = int(dup_mask.sum())
    details['duplicates'] = df[dup_mask].index.tolist()

    # Column cardinality / suspicious constant columns
    cardinality = {col: int(df[col].nunique(dropna=False)) for col in df.columns}
    const_cols = [c for c, v in cardinality.items() if v <= 1]
    near_const = [c for c, v in cardinality.items() if v <= max(2, int(0.01 * n)) and c not in const_cols]
    details['cardinality'] = {'unique_counts': cardinality, 'constant_columns': const_cols, 'near_constant_columns': near_const}

    # Type inconsistencies
    type_issues = {}
    for col in df.columns:
        ser = df[col]
        # If object dtype, look for non-string values
        if ser.dtype == 'O':
            non_null = ser.dropna()
            sample_types = non_null.map(lambda x: type(x).__name__).value_counts().to_dict()
            if len(sample_types) > 1:
                type_issues[col] = sample_types
        # If numeric-like stored as object, try coercion
        if ser.dtype == 'O':
            coerced = pd.to_numeric(ser, errors='coerce')
            coerced_na = coerced.isna() & ~ser.isna()
            if coerced_na.any():
                type_issues.setdefault(col, {})['non_numeric_count_when_coerce'] = int(coerced_na.sum())
    details['type_issues'] = type_issues

    # Date parsing issues: try to auto-parse any column with 'date' in the name or dtype object
    date_parsing = {}
    potential_date_cols = [c for c in df.columns if 'date' in c.lower() or 'time' in c.lower()]
    potential_date_cols += [c for c in df.columns if df[c].dtype == 'O' and df[c].dropna().shape[0] > 0 and df[c].astype(str).str.match(r"\d{4}-\d{2}-\d{2}").any()]
    potential_date_cols = list(dict.fromkeys(potential_date_cols))
    for c in potential_date_cols:
        ser = df[c]
        fmt = None
        if isinstance(date_formats, dict) and c in date_formats:
            fmt = date_formats[c]
        parsed = pd.to_datetime(ser, format=fmt, errors='coerce', infer_datetime_format=(fmt is None))
        bad = parsed.isna() & ~ser.isna()
        if bad.any():
            date_parsing[c] = int(bad.sum())
    details['date_parsing'] = date_parsing

    # Numeric outliers: IQR and z-score
    num_cols = df.select_dtypes(include=[np.number]).columns.tolist()
    outliers = {'iqr': {}, 'zscore': {}}
    for c in num_cols:
        ser = df[c].dropna().astype(float)
        if ser.shape[0] < 5:
            continue
        q1 = ser.quantile(0.25)
        q3 = ser.quantile(0.75)
        iqr = q3 - q1
        lo = q1 - iqr_multiplier * iqr
        hi = q3 + iqr_multiplier * iqr
        iqr_mask = (df[c] < lo) | (df[c] > hi)
        if iqr_mask.any():
            outliers['iqr'][c] = int(iqr_mask.sum())
        # zscore
        if ser.std(ddof=0) == 0:
            z_mask = pd.Series([False] * len(df))
        else:
            z = np.abs(stats.zscore(ser, nan_policy='omit'))
            z_full = pd.Series(index=ser.index, data=z)
            z_mask = z_full > zscore_thresh
        if z_mask.any():
            outliers['zscore'][c] = int(z_mask.sum())
    details['outliers'] = outliers

    # ID-like columns validation (columns ending with Id/ID or containing 'id')
    import re
    id_cols = [c for c in df.columns if re.search(r'id$|_id$|id_', c, flags=re.I) or 'id' == c.lower()]
    id_issues = {}
    for c in id_cols:
        ser = df[c].dropna().astype(str)
        invalid_mask = ~ser.str.match(id_regex)
        if invalid_mask.any():
            id_issues[c] = int(invalid_mask.sum())
    details['id_issues'] = id_issues

    # High cardinality object columns (useful for spotting free-text vs categorical)
    high_card = [c for c in df.select_dtypes(include=['object']).columns if df[c].nunique(dropna=False) > 0.5 * n]
    details['high_cardinality_objects'] = high_card

    # Build a one-row summary DataFrame
    summary = {
        'rows': n,
        'columns': df.shape[1],
        'total_missing_cells': int(df.isnull().sum().sum()),
        'columns_with_missing': int((df.isnull().sum() > 0).sum()),
        'duplicate_rows': int(dup_count),
        'constant_columns': len(const_cols),
        'near_constant_columns': len(near_const),
        'num_numeric_columns': len(num_cols),
        'num_id_like_columns': len(id_cols),
        'potential_date_columns_checked': len(potential_date_cols),
        'columns_with_type_issues': len(type_issues)
    }
    summary_df = pd.DataFrame([summary])

    # Attach some quick sample outputs for immediate inspection (first few problem rows for key issues)
    sample_outputs = {}
    if details['missing']:
        top_missing = missing_summary.sort_values('missing_pct', ascending=False).head(5)
        sample_outputs['top_missing_columns'] = top_missing.to_dict(orient='index')
    if details['date_parsing']:
        sample_outputs['date_parsing'] = details['date_parsing']
    if details['type_issues']:
        sample_outputs['type_issues'] = details['type_issues']
    if details['outliers']['iqr']:
        sample_outputs['outliers_iqr'] = details['outliers']['iqr']

    details['summary'] = summary
    details['sample_outputs'] = sample_outputs

    # Print a concise human-readable summary for interactive use
    print("Data quality quick summary:\n")
    print(summary_df.T)
    if details['missing']:
        print('\nTop missing columns (count, pct):')
        print(missing_summary.sort_values('missing_pct', ascending=False).head(10))
    if details['date_parsing']:
        print('\nDate parsing problems (col: bad_count):')
        print(details['date_parsing'])
    if details['type_issues']:
        print('\nColumns with mixed types or coercion failures:')
        print(details['type_issues'])
    if details['outliers']['iqr'] or details['outliers']['zscore']:
        print('\nOutliers detected (sample):')
        print(details['outliers'])
    if details['duplicates']:
        print(f"\nDuplicate rows detected: {len(details['duplicates'])} rows (use df[df.duplicated(keep=False)] to inspect)")
    if details['id_issues']:
        print('\nID format issues (col: invalid_count):')
        print(details['id_issues'])

    return summary_df, details

# Example usage (uncomment to run):
report, details = detect_errors(flood_project_df)
display(report)
details.keys()
```
